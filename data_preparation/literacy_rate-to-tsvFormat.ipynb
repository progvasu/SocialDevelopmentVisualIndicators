```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv("../data/literacy_rate_raw.csv", skiprows=4)
raw_data.head()
```

```python
raw_data.drop(columns=['Indicator Name', 'Country Code', 'Indicator Code', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', 
               '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', 
               '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', 
                       '2019', 'Unnamed: 64'], inplace=True)
```

```python
raw_data.dropna().shape
```

```python
# since we need few countries only - dropping those
raw_data.dropna(inplace=True)
```

```python
raw_data["Country Name"]
```

```python
# dropping rows that are of no interest
countries_to_drop = ['Early-demographic dividend', 'Europe & Central Asia (excluding high income)', 'IDA total',
                    'IDA blend', 'Latin America & Caribbean (excluding high income)', 'Lower middle income',
                     'Late-demographic dividend', 'Middle East & North Africa (excluding high income)', 'Pre-demographic dividend',
                    'Other small states', 'Sub-Saharan Africa (excluding high income)', 'Small states', 'East Asia & Pacific (excluding high income)',
                    'East Asia & Pacific (IDA & IBRD countries)', 'Europe & Central Asia (IDA & IBRD countries)', 'Latin America & the Caribbean (IDA & IBRD countries)',
                    'Middle East & North Africa (IDA & IBRD countries)', 'South Asia (IDA & IBRD)', 'Sub-Saharan Africa (IDA & IBRD countries)'] 

processed_data = raw_data.drop(raw_data[raw_data["Country Name"].isin(countries_to_drop)].index)
```

```python
# rename column
processed_data.rename(columns={'Country Name':'date'}, inplace=True)
```

```python
processed_data = processed_data.T
```

```python
processed_data.to_csv("../data/literacy_rate_final.tsv", sep="\t", header=None)
```
