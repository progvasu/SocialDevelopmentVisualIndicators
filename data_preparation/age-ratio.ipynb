```python
countries = ["South Africa", "Saudi Arabia", "Nepal", "Mexico", "Kenya", "Iraq", "China", "Japan", "India", "Pakistan"]
```

```python
filtered_data = data.drop(data[~data["Country Name"].isin(countries)].index)
```

```python
new_data = filtered_data.melt(id_vars=["Country Name"], 
        var_name="time", 
        value_name="p")
```

```python
new_data = new_data.sort_values(["Country Name", "time"])
```

```python
new_data.rename(columns={"Country Name": "activity"}, inplace=True)
```
